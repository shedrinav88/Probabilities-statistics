**Задание 1**

Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
salaries = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
```

среднее арифметическое:

```python
sum_salaries = 0

for salary in salaries:
  sum_salaries += salary
mean_salaries = sum_salaries/len(salaries)
mean_salaries
```

среднее квадратичное отклонение:

```python
import numpy as np

deviation = 0
for salary in salaries:
  deviation += (salary - mean_salaries)**2

std_salaries = np.sqrt(deviation / len(salaries))
std_salaries
```

смещенная дисперсия:

```python
variance = std_salaries**2
variance
```

несмещенная дисперсия:

```python
unbiased_std = np.sqrt(deviation / (len(salaries) - 1))

unbiased_variance = unbiased_std**2
unbiased_variance
```

в данной ситуации 3 белых мяча могут быть, в следующих ситуациях:
1) из первой корзины достали 0 белых,из второй - 3;
2) из первой корзины - 1 белый, из второй - 2;
3) из первой корзины достали 2 белых, из второй - 1

```python
from math import factorial
```

Вычислим вероятности исходов для первой козины:

```python
p1_0 = (3/8)**2
p1_1 = ((factorial(2)/((factorial(1)*factorial(1))))*pow(5/8, 1)*pow(3/8, 1))
p1_2 = (5/8)**2
```

Вычислим вероятности исходов для второй козины:

```python
# p2_0 = ((factorial(4)/((factorial(0)*factorial(4))))*pow(5/12, 0)*pow(7/12, 4))
p2_1 = ((factorial(4)/((factorial(1)*factorial(3))))*pow(5/12, 1)*pow(7/12, 3))
p2_2 = ((factorial(4)/((factorial(2)*factorial(2))))*pow(5/12, 2)*pow(7/12, 2))
p2_3 = ((factorial(4)/((factorial(3)*factorial(1))))*pow(5/12, 3)*pow(7/12, 1))
```

Вероятность того, что выпадет ровно 3 белых шара:

```python
prob = p1_0*p2_3 + p1_1*p2_2 + p1_2*p2_1
prob
```
